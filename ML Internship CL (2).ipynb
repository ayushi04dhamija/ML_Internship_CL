```python
df['Close']=df['Close Price']
```

```python
df['Day_Perc_Change']=df.Close.pct_change()* 100
```

```python
df.head(5)
```

```python
def f(row):
    if row['Day_Perc_Change']>=-0.5 and row['Day_Perc_Change']<=0.5:
        return 'Slight or No change'
    if row['Day_Perc_Change']>=0.5 and row['Day_Perc_Change']<=1:
        return 'Slight positive'
    if row['Day_Perc_Change']>=-0.5 and row['Day_Perc_Change']<=-1:
        return 'Slight negative'
    if row['Day_Perc_Change']>=1 and row['Day_Perc_Change']<=3:
        return 'Positive'
    if row['Day_Perc_Change']>=-3 and row['Day_Perc_Change']<=-1:
        return 'Negative'
    if row['Day_Perc_Change']>=3 and row['Day_Perc_Change']<=7:
        return 'Among top gainers'
    if row['Day_Perc_Change']>=-7 and row['Day_Perc_Change']<=-3:
        return 'Among top losers'
    if row['Day_Perc_Change']>=7:
    
        return 'Bull run'
    if row['Day_Perc_Change']<=-7:
        return 'Bear drop'
            
    
    
    
    
    
    
   
```

```python
#df.assign(Trends=[if Day_Perc_Change(-0.5,0.5) =='Slight or No change'])
```

```python
#df['Trend'] = df.apply(f, axis=1)
df['Trend']=df.apply (lambda row: f(row), axis=1)
```
